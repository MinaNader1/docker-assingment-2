```python
import pandas as pd
```

```python
df = pd.read_csv('books.csv')
df
```

# Data Cleaning

```python
df.dropna(inplace=True)
df.drop_duplicates(inplace=True)
df
```

# Data Preprocessing

```python
df = df[df['original_title'].str.contains('Harry Potter', case=False)]
columns_to_drop = [
    'book_id', 'goodreads_book_id', 'best_book_id', 'work_id', 
    'books_count', 'isbn', 'isbn13', 'work_ratings_count', 
    'work_text_reviews_count', 'image_url', 'small_image_url', 'language_code'
]

df.drop(columns_to_drop, axis=1, inplace=True)
df
```

# the most selling books within the Harry Potter series

```python
most_selling_books = df.sort_values(by="ratings_count", ascending=False)
most_selling_books[["original_title", "authors", "ratings_count"]]
```

# the average rating of the Harry Potter books

```python
average_rating = df['average_rating'].mean()
average_rating
```
